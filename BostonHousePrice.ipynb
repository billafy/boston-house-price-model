```python
import numpy as np # importing required libraries
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt

from sklearn.datasets import load_boston # loading the features and target from the boston dataset

load_boston = load_boston()

X = load_boston.data
y = load_boston.target

data = pd.DataFrame(X, columns = load_boston.feature_names)
data['SalePrice'] = y
data.head()
```

```python
sns.displot(data['SalePrice'])
```

```python
skewness = data['SalePrice'].skew()
kurtosis = data['SalePrice'].kurt()
print(f'Skewness : {skewness}\nKurtosis : {kurtosis}')
```

```python
fig, ax = plt.subplots()
ax.scatter(x = data['CRIM'], y = data['SalePrice'])
plt.xlabel('Crime Rate per Capita')
plt.ylabel('House Sale Price')
plt.show()
```

```python
from scipy import stats
from scipy.stats import norm, skew

sns.distplot(data['SalePrice'], fit=norm)

(mu, sigma) = norm.fit(data['SalePrice'])

print(f'Mean : {mu}\nStandard Deviation : {sigma}')

fig = plt.figure()
res = stats.probplot(data['SalePrice'], plot=plt)
plt.show()
```

```python
plt.figure(figsize = (10, 10))
cor = data.corr()
sns.heatmap(cor, annot = True, cmap = plt.cm.PuBu)

plt.show()
```

```python
salePriceCor = abs(cor['SalePrice'])
relevantFeatures = salePriceCor[salePriceCor > 0.35]
names = [name for name, value in relevantFeatures.iteritems()] 

names.pop()

print(names)
```

```python
from sklearn.model_selection import train_test_split

X = data[names]
y = data['SalePrice']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 42)
```

```python
print(f'Number of examples for training : {len(X_train)}\nNumber of examples for testing : {len(X_test)}')
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
predictions = lr.predict(X_test)

print(f'Actual price of the house : {y_test[0]}\nPredicted price of the house : {predictions[0]}')
```

```python
differences = y_test - predictions
correct = len(differences[abs(differences) <= 5])
efficiency = correct * 100 / len(y_test)
        
print(f'Efficiency of the model : {efficiency}%')
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test, predictions)
rmse = np.sqrt(mse)

print(f'Cost : {mse}')
```
